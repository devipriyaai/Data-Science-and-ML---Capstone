## Importing the Dataset

```python
dataset = pd.read_csv("Rainfall.csv")
```

```python
# Print data types of all columns
print(dataset.dtypes)
```

```python
# Strip any leading/trailing spaces from column names
dataset.columns = dataset.columns.str.strip()

# Verify the cleaned column names
print(dataset.columns)
```

```python
dataset.isna().sum()
```

```python
dataset['winddirection'].fillna(dataset['winddirection'].median(), inplace=True)
dataset['windspeed'].fillna(dataset['windspeed'].median(), inplace=True)
```

## Univarite Analysis

```python
import seaborn as sns
```

```python
sns.distplot(dataset['temparature'],hist=True,kde=True,kde_kws={'color':'red'})
```

## Bivariate Analysis:Distribution Plot

```python
sns.jointplot(x='temparature',y='humidity',data=dataset)
plt.show()
```

```python
sns.jointplot(x='day',y='sunshine',data=dataset,kind='hex')
plt.show()
```

```python
sns.jointplot(x='maxtemp',y='mintemp',data=dataset,kind='kde')
plt.show()
```

```python
# Pair plot
```

```python
sns.pairplot(dataset,hue='day',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

## For Categorical Plot

```python
sns.stripplot(x='day',y='cloud',data=dataset)
plt.show()
```
