```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset= pd.read_csv("preprocessed_Rainfall_dataset.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
# Step 1: Convert 'rainfall' to binary (if not already done)
dataset['rainfall'] = dataset['rainfall'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype =int,drop_first = True)
```

```python
dataset
```

```python
dataset["rainfall"].value_counts()
```

```python
dataset.columns
```

```python
independent = dataset[['day', 'pressure', 'maxtemp', 'temparature', 'mintemp', 'dewpoint',
                'humidity', 'cloud','sunshine', 'winddirection','windspeed']]
```

```python
independent
```

```python
dependent = dataset[["rainfall"]]
```

```python
dependent
```

```python
independent.shape
```

```python
dependent.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors =7,metric = 'minkowski',p =2)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report= classification_report(y_test,y_pred)
```

```python
print(clf_report)
```
